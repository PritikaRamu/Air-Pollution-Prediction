```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Alcantarilla (18-19)

```python
df_18 = pd.read_csv('Alcantarillahorarias2018.csv', header=1)
df_18= df_18.drop(['PRB (mb)','C6H6 (µg/m³N)','C7H8 (µg/m³N)','XIL (µg/m³N)'],axis =1)
df_18.rename(columns = {'Fecha':'Date'}, inplace = True) 
df_18
```

```python
plt.figure(figsize=(16, 6))
sns.heatmap(df_18.corr(method='pearson'), vmin=-1, vmax=1, annot=True)
```

# Imputing missing values

```python
alcantarilla['Date'] = pd.to_datetime(alcantarilla['Date'], format='%d-%m-%Y')
alcantarilla = alcantarilla.set_index('Date')
alcantarilla.info()
```

## Various interpolation methods

```python
alcantarilla_linear = alcantarilla.assign(InterpolateLinear=alcantarilla['SO2 (µg/m³N)'].interpolate(method='linear'))
alcantarilla_time = alcantarilla.assign(InterpolateTime=alcantarilla['SO2 (µg/m³N)'].interpolate(method='time'))
alcantarilla_quadratic = alcantarilla.assign(InterpolateQuadratic=alcantarilla['SO2 (µg/m³N)'].interpolate(method='quadratic'))
alcantarilla_cubic = alcantarilla.assign(InterpolateCubic=alcantarilla['SO2 (µg/m³N)'].interpolate(method='cubic'))
alcantarilla_slinear = alcantarilla.assign(InterpolateSLinear=alcantarilla['SO2 (µg/m³N)'].interpolate(method='slinear'))
alcantarilla_akima = alcantarilla.assign(InterpolateAkima=alcantarilla['SO2 (µg/m³N)'].interpolate(method='akima'))
alcantarilla_poly5 = alcantarilla.assign(InterpolatePoly5=alcantarilla['SO2 (µg/m³N)'].interpolate(method='polynomial', order=5)) 
alcantarilla_poly7 = alcantarilla.assign(InterpolatePoly7=alcantarilla['SO2 (µg/m³N)'].interpolate(method='polynomial', order=7))
alcantarilla_spline3 = alcantarilla.assign(InterpolateSpline3=alcantarilla['SO2 (µg/m³N)'].interpolate(method='spline', order=3))
alcantarilla_spline4 = alcantarilla.assign(InterpolateSpline4=alcantarilla['SO2 (µg/m³N)'].interpolate(method='spline', order=4))
alcantarilla_spline5 = alcantarilla.assign(InterpolateSpline5=alcantarilla['SO2 (µg/m³N)'].interpolate(method='spline', order=5))
```

## kNN imputation

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5) 
After_imputation = imputer.fit_transform(alcantarilla)
alc_knn = pd.DataFrame(After_imputation, columns = ['NO (µg/m³N)','NO2 (µg/m³N)','SO2 (µg/m³N)','O3 (µg/m³N)','TMP (ºC)','HR (% H.R.)','NOX (µg/m³N)','DD (grados)','RS (W/m³)','VV (m/s)','PM10 (µg/m³N)'])
alc_knn
```

```python
fig, axs = plt.subplots(4, 3)

alc_knn['Date'] = d[:-1]

axs[0, 0].plot(alc_knn['Date'],alc_knn['NO (µg/m³N)'])
axs[0, 0].set_title('NO (µg/m³N)')

axs[0, 1].plot(alc_knn['Date'],alc_knn['NO2 (µg/m³N)'], 'tab:orange')
axs[0, 1].set_title('NO2 (µg/m³N)')

axs[0, 2].plot(alc_knn['Date'],alc_knn['SO2 (µg/m³N)'], 'tab:green')
axs[0, 2].set_title('SO2 (µg/m³N)')

axs[1, 0].plot(alc_knn['Date'],alc_knn['O3 (µg/m³N)'], 'tab:red')
axs[1, 0].set_title('O3 (µg/m³N)')

axs[1, 1].plot(alc_knn['Date'],alc_knn['NOX (µg/m³N)'], 'tab:purple')
axs[1, 1].set_title('NOX (µg/m³N)')

axs[1, 2].plot(alc_knn['Date'],alc_knn['HR (% H.R.)'])
axs[1, 2].set_title('HR (% H.R.)')

axs[2, 0].plot(alc_knn['Date'],alc_knn['DD (grados)'])
axs[2, 0].set_title('DD (grados)')

axs[2, 1].plot(alc_knn['Date'],alc_knn['RS (W/m³)'], 'tab:orange')
axs[2, 1].set_title('RS (W/m³)')

axs[2, 2].plot(alc_knn['Date'],alc_knn['PM10 (µg/m³N)'], 'tab:green')
axs[2, 2].set_title('PM10 (µg/m³N)')

axs[3, 0].plot(alc_knn['Date'],alc_knn['TMP (ºC)'], 'tab:green')
axs[3, 0].set_title('TMP (ºC)')

axs[3, 1].plot(alc_knn['Date'],alc_knn['VV (m/s)'], 'tab:purple')
axs[3, 1].set_title('VV (m/s)')

for i in [0,1,2,3]:
    for j in [0,1,2]:
        axs[i, j].tick_params(axis='x',which='both', bottom=False, top=False, labelbottom=False)
```
