```python
fig, axs = plt.subplots(3, 3)

axs[0, 0].plot(df['Fecha'],df['NO (µg/m³N)'])
axs[0, 0].set_title('NO (µg/m³N)')

axs[0, 1].plot(df['Fecha'],df['NO2 (µg/m³N)'], 'tab:orange')
axs[0, 1].set_title('NO2 (µg/m³N)')

axs[0, 2].plot(df['Fecha'],df['SO2 (µg/m³N)'], 'tab:green')
axs[0, 2].set_title('SO2 (µg/m³N)')

axs[1, 0].plot(df['Fecha'],df['O3 (µg/m³N)'], 'tab:red')
axs[1, 0].set_title('O3 (µg/m³N)')

axs[1, 1].plot(df['Fecha'],df['NOX (µg/m³N)'], 'tab:purple')
axs[1, 1].set_title('NOX (µg/m³N)')

axs[1, 2].plot(df['Fecha'],df['C6H6 (µg/m³N)'])
axs[1, 2].set_title('C6H6 (µg/m³N)')

axs[2, 0].plot(df['Fecha'],df['C7H8 (µg/m³N)'])
axs[2, 0].set_title('C7H8 (µg/m³N)')

axs[2, 1].plot(df['Fecha'],df['XIL (µg/m³N)'], 'tab:orange')
axs[2, 1].set_title('XIL (µg/m³N)')

axs[2, 2].plot(df['Fecha'],df['PM10 (µg/m³N)'], 'tab:green')
axs[2, 2].set_title('PM10 (µg/m³N)')

for i in [0,1,2]:
    for j in [0,1,2]:
        axs[i, j].tick_params(axis='x',which='both', bottom=False, top=False, labelbottom=False)
```
